```python
# 11개 변수의 정규분포 확인
from scipy.stats import shapiro, normaltest
import statsmodels.api as sm
import warnings

warnings.filterwarnings("ignore")

non_normal_columns = []

for column in new_df.columns:
    data = new_df[column]
    
    shapiro_result = shapiro(data)
    normaltest_result = normaltest(data)
    
    if shapiro_result.pvalue < 0.05 or normaltest_result.pvalue < 0.05:
        non_normal_columns.append(column)
        print(f'Column: {column}')
        print(f'Shapiro-Wilk test p-value: {shapiro_result.pvalue}')
        print(f'Normality test p-value: {normaltest_result.pvalue}')
        
        # Create subplots for visualizations
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        # Plot histogram
        sns.histplot(data, ax=axes[0], kde=True)
        axes[0].set_title(f'Histogram of {column}')
        
        # Q-Q plot
        axes[1] = plt.subplot(122)
        stats_probplot = sm.ProbPlot(data, fit=True)
        stats_probplot.ppplot(line='45', ax=axes[1])
        axes[1].set_title(f'Q-Q plot of {column}')
        plt.tight_layout()
        
        plt.show()
```

# PCA

```python
# 주성분 개수에 따른 누적 분산 비율(설명력) 계산
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# PCA 객체를 생성합니다.
pca = PCA()

# 주성분을 추출하기 전, 데이터를 StandardScaler를 사용하여 정규화합니다.
X_scaled = StandardScaler().fit_transform(new_df)

# 주성분 분석을 수행합니다.
pca.fit(X_scaled)

# 각 주성분이 설명하는 분산의 비율을 계산합니다.
explained_variance_ratio = pca.explained_variance_ratio_

# 누적 설명 분산을 계산합니다.
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 누적 설명 분산 비율을 그래프로 나타냅니다.
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()

```

```python
# 정규화 후 11개 변수의 상관관계 확인
X_scaled_df = pd.DataFrame(X_scaled, columns=['salary_avg', 'house_total','business_total', 'sales_sum','transport_sum','bus_sum',
              'workers_total', 'popLiv_total','popReg_total','business_total.1','car_total'])

correlation_matrix = X_scaled_df.corr()

plt.figure(figsize=(20, 10))

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)

plt.title("Correlation Coefficient Heatmap")
plt.show()
```

```python
# 설명력이 70% 이상인 컴포넌트 개수 출력
components_needed = np.argmax(cumulative_variance_ratio >= 0.7) + 1
print(f"Number of components needed for 80% explained variance: {components_needed}")
```

```python
# 2개의 주성분으로 PCA 수행

# PCA 객체를 생성합니다.
pca = PCA(n_components=2)
# 주성분을 추출하기 전, 데이터를 StandardScaler를 사용하여 정규화합니다.
X_scaled = StandardScaler().fit_transform(new_df)

# 주성분 분석을 수행합니다.
pca.fit(X_scaled)

# 주성분 변환을 수행합니다.
X_pca = pca.transform(X_scaled)

# 변환된 데이터의 크기를 출력합니다.
print('Transformed data shape:', X_pca.shape)

# 변환된 데이터의 처음 5개 샘플을 출력합니다.
print('Transformed data:\n', X_pca[:5])

# 설명된 분산의 비율을 확인합니다.
explained_variance_ratio = pca.explained_variance_ratio_
print('Explained variance ratio:', explained_variance_ratio)

# 누적 설명 분산 비율을 계산합니다.
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 누적 설명 분산 비율을 그래프로 나타냅니다.
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.xlim(0,3.0)
plt.show()
```

```python
# 2개의 주성분 각각의 원본 변수(11개) 기여도 계산

loadings = pca.components_

# 내림차순 정렬
top_features_pc1_indices = np.argsort(loadings[0])[-15:][::-1]
top_features_pc2_indices = np.argsort(loadings[1])[-15:][::-1]

top_features_pc1 = new_df.columns[top_features_pc1_indices]
top_features_pc2 = new_df.columns[top_features_pc2_indices]

print('Top 5 features for PC1:')
for feature, loading in zip(top_features_pc1, loadings[0][top_features_pc1_indices]):
    print(f'{feature}: {loading:.4f}')

print('\nTop 5 features for PC2:')
for feature, loading in zip(top_features_pc2, loadings[1][top_features_pc2_indices]):
    print(f'{feature}: {loading:.4f}')
```

# K-Means

```python
# 최적 클러스터 수 결정을 위한 실루엣 계수와 엘보우 포인트 그래프 생성
# 실루엣 계수(파란색)이 높고 엘보우 포인트(빨간색)가 급격하게 완만해지는 곳이 최적 클러스터 수
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

sse = []
silhouette_coefficients = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    sse.append(kmeans.inertia_)
    
    score = silhouette_score(X_pca, kmeans.labels_)
    silhouette_coefficients.append(score)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(k_range, sse, 'bx-')
ax.set_xlabel('Number of Clusters (k)')
ax.set_ylabel('Sum of squared distance', color='r')
ax.set_title('Elbow Point and Silhouette Score')
ax.tick_params('y', colors='r')

ax2 = ax.twinx()
ax2.plot(k_range, silhouette_coefficients, 'ro-')
ax2.set_ylabel('Silhouette score', color='b')
ax2.tick_params('y', colors='b')

plt.show()

elbow_point = sse.index(min(sse)) + 2
max_silhouette_score = silhouette_coefficients.index(max(silhouette_coefficients)) + 2

print(f"Elbow point number of clusters: {elbow_point}")
print(f"Maximum silhouette score number of clusters: {max_silhouette_score}")
```

```python
# 6개의 클러스터로 k-means 수행 후 시각화, X는 중심점

n_clusters = 6
n_components = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)
cluster_centers = kmeans.cluster_centers_

pca_df = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, n_components+1)])
pca_df['Cluster'] = cluster_labels

distances = np.sqrt(np.sum((X_pca - cluster_centers[cluster_labels])**2, axis=1))
pca_df['Distance'] = distances

closest_points = []
for cluster_id in range(n_clusters):
    cluster_mask = pca_df['Cluster'] == cluster_id
    closest_indices = pca_df[cluster_mask].sort_values(by='Distance').index[:5]
    closest_points.extend(closest_indices)

for cluster_id in range(n_clusters):
    print(f"Cluster {cluster_id} - Closest points:", pca_df.loc[closest_points][pca_df['Cluster'] == cluster_id].index)

color_map = plt.cm.get_cmap('rainbow', n_clusters)

# Visualization
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster'], cmap=color_map)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=100, c='black', label='Cluster Centers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA and K-Means Clustering Visualization')

# Add color legend
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map(cluster_id), markersize=10, label=f'Cluster {cluster_id}') for cluster_id in range(n_clusters)]
plt.legend(handles=legend_handles)

plt.show()
```

# GMM

```python
# 최적 클러스터링 수 결정을 위한 AIC,BIC 그래프 출력
# AIC와 BIC 모두 가장 작은 곳이 최적 클러스터링 수
from sklearn.mixture import GaussianMixture

n_components_range = range(1, 11)
aic_values = []
bic_values = []
for n_components in n_components_range:
    model = GaussianMixture(n_components=n_components, random_state=0)
    model.fit(X_pca)
    aic = model.aic(X_pca)
    bic = model.bic(X_pca)
    aic_values.append(aic)
    bic_values.append(bic)

optimal_aic_clusters = n_components_range[np.argmin(aic_values)]
optimal_bic_clusters = n_components_range[np.argmin(bic_values)]

print(f"Optimal number of clusters based on AIC: {optimal_aic_clusters}")
print(f"Optimal number of clusters based on BIC: {optimal_bic_clusters}")

plt.plot(n_components_range, aic_values, label='AIC')
plt.plot(n_components_range, bic_values, label='BIC')

plt.axvline(x=optimal_aic_clusters, color='red', linestyle='--', label='Optimal AIC')
plt.axvline(x=optimal_bic_clusters, color='blue', linestyle='--', label='Optimal BIC')

plt.title('AIC/BIC Scores for Gaussian Mixture Model')
plt.xlabel('Number of Components')
plt.ylabel('Score')
plt.legend()

plt.show()
```

```python
# GMM 모델을 생성하고 적합합니다.
n_components = 5  # 선택한 클러스터 개수
gmm = GaussianMixture(n_components=n_components, random_state=0)
gmm.fit(X_pca)  

cluster_labels = gmm.predict(X_pca)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='rainbow')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
# 실루엣 계수로 성능 확인
from sklearn.metrics import silhouette_samples, silhouette_score

silhouette_avg = silhouette_score(X_pca, cluster_labels)
sample_silhouette_values = silhouette_samples(X_pca, cluster_labels)

for cluster_label in np.unique(cluster_labels):
    cluster_silhouette_values = sample_silhouette_values[cluster_labels == cluster_label]
    cluster_avg_silhouette = np.mean(cluster_silhouette_values)
    
    print(f"Cluster {cluster_label}: Avg Silhouette Score = {cluster_avg_silhouette:.2f}")
```

```python
from sklearn_extra.cluster import KMedoids

# k-medoids 알고리즘 적용
elbow_scores = []
silhouette_scores = []
for n_clusters in range(2, 7):
    kmedoids = KMedoids(n_clusters=n_clusters).fit(X_pca)
    elbow_scores.append(kmedoids.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, kmedoids.labels_))

# 엘보우 포인트와 실루엣 값을 그래프에 시각화
fig, ax1 = plt.subplots()

# 엘보우 포인트 그래프
ax1.plot(range(2, 7), elbow_scores, 'o-', color='r')
ax1.set_xlabel('Number of clusters')
ax1.set_ylabel('Sum of squared distance', color='r')
ax1.tick_params('y', colors='r')

# 실루엣 값 그래프
ax2 = ax1.twinx()
ax2.plot(range(2, 7), silhouette_scores, 'o-', color='b')
ax2.set_ylabel('Silhouette score', color='b')
ax2.tick_params('y', colors='b')

plt.title('Elbow Point and Silhouette Score')
plt.show()
```

```python
# 6개의 클러스터로 클러스터링 수행  
kmedoids = KMedoids(n_clusters=6, random_state=42)
cluster_labels = kmedoids.fit_predict(X_pca)

pca_df = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, 2+1)])
pca_df['Cluster'] = cluster_labels

plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster'], cmap='rainbow')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA and K-Medoids Clustering Visualization')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# 실루엣 계수로  성능 확인
from sklearn.metrics import silhouette_samples, silhouette_score

silhouette_avg = silhouette_score(pca_df, cluster_labels)
sample_silhouette_values = silhouette_samples(pca_df, cluster_labels)

for cluster_label in np.unique(cluster_labels):
    cluster_silhouette_values = sample_silhouette_values[cluster_labels == cluster_label]
    cluster_avg_silhouette = np.mean(cluster_silhouette_values)
    
    print(f"Cluster {cluster_label}: Avg Silhouette Score = {cluster_avg_silhouette:.2f}")
```
