# PCA 후 K-means (기각)

```python
# 1차 군집화 결과에서 2번 클러스터 사용
new_df = new_df[new_df['Cluster']==2]
new_df = new_df.iloc[:,4:]
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# PCA 객체를 생성합니다.
pca = PCA()

# 주성분을 추출하기 전, 데이터를 StandardScaler를 사용하여 정규화합니다.
X_scaled = StandardScaler().fit_transform(new_df)

# 주성분 분석을 수행합니다.
pca.fit(X_scaled)

# 각 주성분이 설명하는 분산의 비율을 계산합니다.
explained_variance_ratio = pca.explained_variance_ratio_

# 누적 설명 분산을 계산합니다.
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 누적 설명 분산 비율을 그래프로 나타냅니다.
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()

```

```python
# 설명 비율이 70% 이상인 컴포넌트 개수 출력
components_needed = np.argmax(cumulative_variance_ratio >= 0.7) + 1
print(f"Number of components needed for 80% explained variance: {components_needed}")
```

```python
# PCA 객체를 생성합니다.
pca = PCA(n_components=5)
# 주성분을 추출하기 전, 데이터를 StandardScaler를 사용하여 정규화합니다.
X_scaled = StandardScaler().fit_transform(new_df)

# 주성분 분석을 수행합니다.
pca.fit(X_scaled)

# 주성분 변환을 수행합니다.
X_pca = pca.transform(X_scaled)

# 변환된 데이터의 크기를 출력합니다.
print('Transformed data shape:', X_pca.shape)

# 변환된 데이터의 처음 5개 샘플을 출력합니다.
print('Transformed data:\n', X_pca[:5])

# 설명된 분산의 비율을 확인합니다.
explained_variance_ratio = pca.explained_variance_ratio_
print('Explained variance ratio:', explained_variance_ratio)

# 누적 설명 분산 비율을 계산합니다.
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# 누적 설명 분산 비율을 그래프로 나타냅니다.
plt.plot(cumulative_variance_ratio)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.xlim(0,3.0)
plt.show()
```

```python
# 최적 클러스터 수 결정을 위한 실루엣 계수와 엘보우 포인트 시각화 생성
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

sse = []
silhouette_coefficients = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_pca)
    sse.append(kmeans.inertia_)
    
    score = silhouette_score(X_pca, kmeans.labels_)
    silhouette_coefficients.append(score)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(k_range, sse, 'bx-')
ax.set_xlabel('Number of Clusters (k)')
ax.set_ylabel('Sum of squared distance', color='r')
ax.set_title('Elbow Point and Silhouette Score')
ax.tick_params('y', colors='r')

ax2 = ax.twinx()
ax2.plot(k_range, silhouette_coefficients, 'ro-')
ax2.set_ylabel('Silhouette score', color='b')
ax2.tick_params('y', colors='b')

plt.show()

elbow_point = sse.index(min(sse)) + 2
max_silhouette_score = silhouette_coefficients.index(max(silhouette_coefficients)) + 2

print(f"Elbow point number of clusters: {elbow_point}")
print(f"Maximum silhouette score number of clusters: {max_silhouette_score}")
```

```python
# 6개의 클러스터로 k-means 수행
n_clusters = 6
n_components = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)
cluster_centers = kmeans.cluster_centers_

pca_df = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, n_components+1)])
pca_df['Cluster'] = cluster_labels

distances = np.sqrt(np.sum((X_pca - cluster_centers[cluster_labels])**2, axis=1))
pca_df['Distance'] = distances

closest_points = []
for cluster_id in range(n_clusters):
    cluster_mask = pca_df['Cluster'] == cluster_id
    closest_indices = pca_df[cluster_mask].sort_values(by='Distance').index[:5]
    closest_points.extend(closest_indices)

for cluster_id in range(n_clusters):
    print(f"Cluster {cluster_id} - Closest points:", pca_df.loc[closest_points][pca_df['Cluster'] == cluster_id].index)

color_map = plt.cm.get_cmap('rainbow', n_clusters)

# Visualization
plt.scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['Cluster'], cmap=color_map)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=100, c='black', label='Cluster Centers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA and K-Means Clustering Visualization')

# Add color legend
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color_map(cluster_id), markersize=10, label=f'Cluster {cluster_id}') for cluster_id in range(n_clusters)]
plt.legend(handles=legend_handles)

plt.show()
```

```python
# 실루엣 계수로 성능 확인
from sklearn.metrics import silhouette_samples, silhouette_score

silhouette_avg = silhouette_score(pca_df, cluster_labels)
sample_silhouette_values = silhouette_samples(pca_df, cluster_labels)

for cluster_label in np.unique(cluster_labels):
    cluster_silhouette_values = sample_silhouette_values[cluster_labels == cluster_label]
    cluster_avg_silhouette = np.mean(cluster_silhouette_values)
    
    print(f"Cluster {cluster_label}: Avg Silhouette Score = {cluster_avg_silhouette:.2f}")
    
# 결과가 좋지 못하므로 기각
```

# 1. PCA -> T-SNE

```python
# 높은 설명력을 보이는 컴포넌트 개수로 PCA 후 t-SNE
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# PCA 객체를 생성합니다.
pca = PCA(n_components=14)

# 주성분을 추출하기 전, 데이터를 StandardScaler를 사용하여 정규화합니다.
X_scaled = StandardScaler().fit_transform(data)

# 주성분 분석을 수행합니다.
pca.fit(X_scaled)

# 주성분 변환을 수행합니다.
X_pca = pca.transform(X_scaled)
# 40/300
# perplexity와 n_iter, learning_rate를 바꿔가며 [t-SNE] KL divergence after 400 iterations 수치 확인
# perplexity = 5~50
# n_iter = 250~1000
# learning_rate = 100~1000
# perplexity는 10씩, 나머지는 100씩 변화시키면서 확인
model = TSNE(n_components=2,verbose=1,perplexity=50,n_iter=1000,random_state=190)
transformed = model.fit_transform(X_pca)

plt.rcParams['figure.figsize'] = [10, 8]
xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs,ys)
plt.show() 
```

# 최적 결과를 위한 ramdon_state 결정

```python
from sklearn.manifold import TSNE

# 데이터 및 변수 설정
num_iterations = 200  # 이 값을 원하는 반복 횟수로 변경 가능
min_kl_divergence = float('inf')  # 초기의 최대값 설정
best_random_state = None

# 반복문을 사용하여 여러 random_state 값을 테스트
for i in range(num_iterations):
    model = TSNE(n_components=2, verbose=0, perplexity=50, n_iter=1000, random_state=i)
    model.fit_transform(X_pca)
    
    # 현재 KL divergence 값 확인
    current_kl_divergence = model.kl_divergence_
    
    # 현재 KL divergence 값이 이전의 최소값보다 작은 경우 업데이트
    if current_kl_divergence < min_kl_divergence:
        min_kl_divergence = current_kl_divergence
        best_random_state = i

print(f"Best random_state: {best_random_state}")
print(f"Minimum KL divergence: {min_kl_divergence}")
# 수행 결과 196, 196 주변의 시각적으로 가장 나은 190 결정
# 수행 후 KL Divergence가 더 낮은 2. T-SNE만 수행하는 방법 사용
```

```python
# 최적 eps 구간 확인
# 급격하게 증가하는 구간이 최적 eps 구간
from sklearn.neighbors import NearestNeighbors

knn_model = NearestNeighbors(n_neighbors=4)  
knn_model.fit(transformed)

distances, _ = knn_model.kneighbors(transformed)

sorted_distances = np.sort(distances[:, -1])   
plt.plot(sorted_distances)
plt.title(f'K-nearest Neighbors Distance Plot (K = 4)')
plt.xlabel('Data Point Index')
plt.ylabel('4-Nearest Neighbor Distance')
plt.show()

elbow_index = np.argmax(np.diff(sorted_distances)) + 1
optimal_eps = sorted_distances[elbow_index]

print(f"Optimal epsilon (eps): {optimal_eps}")

model = DBSCAN(eps=optimal_eps, min_samples=11)
predict = model.fit(transformed)
cluster_counts = pd.Series(predict.labels_).value_counts()

print("Cluster Counts:")
print(cluster_counts)
# 최적 eps 0.3으로 결정
```

```python
# 고정된 eps에서, min_samples 수를 변경해가며 클러스터 개수 확인
min_samples_range = range(2, 50)

max_cluster_count = 0
optimal_min_samples = None

for min_samples in min_samples_range:
    db_model = DBSCAN(eps=0.3, min_samples=min_samples)
    predict = db_model.fit(transformed)
    cluster_labels = pd.Series(predict.labels_)
    cluster_count = cluster_labels.nunique()
    
    if cluster_count > max_cluster_count:
        max_cluster_count = cluster_count
        optimal_min_samples = min_samples

    print(f"min_samples = {min_samples}, Cluster Count = {cluster_count}")

print(f"Optimal min_samples for the largest number of clusters: {optimal_min_samples}")
# t-sne는 DBSCAN의 거리 방식과 다르게 나타날 수 있음
# 보수적으로 접근해 낮은 min_samples 수와 많은 클러스터 수로 접근
# 2로 결정
```
