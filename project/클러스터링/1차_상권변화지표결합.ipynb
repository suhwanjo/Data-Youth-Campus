```python
import pandas as pd
import numpy as np
```

# 데이터 불러오기 + 사용할 년도 지정

```python
df = pd.read_csv('서울시 상권분석서비스(행정동별 상권변화지표).csv',encoding='cp949')
df = df[df['기준_년_코드']==2015] # 산업 관련 데이터가 대부분 2014~2015데이터 였기 때문에 2015년 1분기 선택
df = df[df['기준_분기_코드']==1]
df
```

```python
# 클러스터 결과 데이터 불러오기
cl = pd.read_csv('kmeans_총합_소득수준_아파트수정.csv')
cl
```

# 행정동 맞추기

```python
a = set(df['행정동_코드_명'])
b = set(cl['Dong'])
a^b
```

```python
no_dong = ['상일동','위례동','일원2동']

# 위의 리스트를 제외한 데이터만 추출
df = df[~(df['행정동_코드_명'].isin(no_dong))]
df
```

# 신사동 나누기

```python
df[df['행정동_코드_명']=='신사동']
```

```python
cl[cl['Dong']=='신사동_관']
```

```python
# 신사동 식별을 위한 이름 부여
df.loc[df['행정동_코드'] == 11620685, '행정동_코드_명'] = "신사동_관"
df.loc[df['행정동_코드'] == 11680510, '행정동_코드_명'] = "신사동_강"
```

```python
a = set(df['행정동_코드_명'])
b = set(cl['Dong'])
a^b
```

```python
# 클러스터 결과 데이터와 상권변화지표 결합
df1 = pd.merge(df,cl,how='inner',left_on='행정동_코드_명',right_on='Dong')
df1 = df1.drop(columns=['행정동_코드','행정동_코드_명'])
df1
```

# 클러스터 별 상권변화지표 집계

```python
# 클러스터 별 상권변화지표 별 count 집계
grouped = df1.groupby(['Cluster', '상권_변화_지표_명']).size().reset_index(name='Count')
grouped
```

# 클러스터별 운영/폐업 영업 개월 평균 집계

```python
# 클러스터 별 운영/폐업 영업개월평균 합 집계
df1.groupby('Cluster').agg(운영영업개월=('운영_영업_개월_평균','sum'),폐업영업개월=('폐업_영업_개월_평균','sum'),
                           서울운영영업개월=('서울_운영_영업_개월_평균','sum'),서울폐업영업개월=('서울_폐업_영업_개월_평균','sum'))
```
