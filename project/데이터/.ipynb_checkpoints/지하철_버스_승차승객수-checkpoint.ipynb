```python
import pandas as pd
import csv
import numpy as np
```

```python
daejung = pd.read_csv("C:/Users/ASUS/Desktop/데이터/서울시 행정동별 대중교통 총 승차 승객수 정보.csv",encoding='cp949')
daejung
```

```python
# Convert the date-like string to datetime format
daejung['기준_날짜'] = pd.to_datetime(daejung['기준_날짜'], format='%Y%m%d')

# Extract rows with December 2022
daejung = daejung[(daejung['기준_날짜'] >= '2022-12-01') & (daejung['기준_날짜'] < '2023-01-01')]
daejung
```

```python
dong = pd.read_csv('C:/Users/ASUS/Desktop/데이터/찐찐행정동420.csv')
dong
```

```python
a = set(daejung['행정동_ID'].unique())
b = set(dong['ADSTRD_CD_x'].unique())

a-b
```

```python
only_daejung_dongs = [1123074, 1124081, 1124082, 1125052]

# 주어진 행정동명을 포함하지 않은 행 추출
daejung = daejung[~daejung['행정동_ID'].isin(only_daejung_dongs)]
```

```python
df = pd.merge(daejung,dong,how='inner',left_on='행정동_ID',right_on='ADSTRD_CD_x')
df
```

```python
a = set(df['ADSTRD_NM'].unique())
b = set(dong['ADSTRD_NM'].unique())
```

```python
df1 = df.groupby('행정동_ID').sum()
```

```python
df1
```

```python
# Calculate column averages
column_averages = df1.mean()

# Add a new row with the column averages
df1.loc['장지동'] = column_averages
```
