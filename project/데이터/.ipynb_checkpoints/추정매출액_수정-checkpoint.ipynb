```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
folder = "C:/Users/ASUS/Desktop/데이터/기타반출자료/B058. 서울시 행정동단위 분기별 추정매출액/2. 파일데이터"
data_list = []

for subfol in os.listdir(folder):
    subfol_path = os.path.join(folder, subfol)
    
    if os.path.isdir(subfol_path):
        for file in os.listdir(subfol_path):
            file_path = os.path.join(subfol_path, file)
            df = pd.read_csv(file_path, encoding='utf-8-sig', sep='|')
            data_list.append(df)
            
money = pd.concat(data_list, ignore_index=True)
money
# 기준년 코드
# 기준 분기 코드
# 행정동 코드
# 행정동명
# 서비스업종코드
# 서비스업종코드명
# 연령대_10세단위
# 매출액_합계
# 매출건수_합계
```

# 행정동 코드 추출

```python
gu = money[['ADSTRD_CD','ADSTRD_NM']]
gu = gu[gu['ADSTRD_NM']!='위례동']
```

```python
# Step 3: Drop duplicate codes
df_unique_codes = gu.drop_duplicates(subset='ADSTRD_CD')

# Step 4: Group by code and check if values are the same
df_unique_codes['consistent_value'] = gu.groupby('ADSTRD_CD')['ADSTRD_NM'].transform('nunique') == 1

# Step 5: Select rows with consistent values
unique_pairs = df_unique_codes[df_unique_codes['consistent_value']][['ADSTRD_CD', 'ADSTRD_NM']].reset_index(drop=True)

# Print the unique code-value pairs
print(unique_pairs)
```

```python
unique_pairs.to_csv("행정동코드8자리.csv",index=False)
```
