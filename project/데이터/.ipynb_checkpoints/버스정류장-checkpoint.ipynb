```python
# KaKao API를 사용해서 좌표 -> 행정동 변환하는 함수
# API키 유출 하지 마셈요
import requests

def get_admin_region_info(x, y, format='json', input_coord='WGS84', output_coord='WGS84', api_key='ea302f77b2a44a9aecc040db94438f56'):
    url = f"https://dapi.kakao.com/v2/local/geo/coord2regioncode.{format}"
    headers = {'Authorization': f'KakaoAK {api_key}'}
    params = {
        'x': str(x),
        'y': str(y),
        'input_coord': input_coord,
        'output_coord': output_coord
    }

    response = requests.get(url, headers=headers, params=params)
    response_json = response.json()

    return response_json

# Kakao API 호출 및 행정동명 추가
def add_admin_region_name(row):
    result = get_admin_region_info(row['좌표X'], row['좌표Y'])
    if 'documents' in result:
        document = result['documents'][1]
        row['행정동명'] = document['region_3depth_name']
        row['시군구명'] = document['region_2depth_name']
    return row
```

```python
df = bus.apply(add_admin_region_name, axis=1)
```

# 신사동 분리

```python
df = pd.read_csv('버스정류장.csv')
df
```

```python
len(df['행정동명'].unique())
```

```python
dong = pd.read_csv('신사동분리버전행정동코드.csv')
dong
```

```python
a = set(df['행정동명'].unique())
b = set(dong['ADSTRD_NM'].unique())
a^b
```

```python
dong_dic = {
    '금호2.3가동':'금호2·3가동',
    '면목3.8동':'면목3·8동',
    '면목제3.8동':'면목3·8동',
    '상계3.4동':'상계3·4동',
    '상계6.7동':'상계6·7동',
    '종로1.2.3.4가동':'종로1·2·3·4가동',
    '종로5.6가동':'종로5·6가동',
    '중계2.3동':'중계2·3동'
}
# Function to apply the mapping
def apply_mapping(value):
    if value in dong_dic:
        return dong_dic[value]
    return value
df['행정동명'] = df['행정동명'].apply(apply_mapping)
```

```python
df1 = pd.merge(df,dong,how='inner',left_on='행정동명',right_on='ADSTRD_NM')
df1
```

```python
df1.to_csv('버스정류장.csv',index=False)
# 421 -> 420 행정동 코드 내보내기
#dong.to_csv('C:/Users/shjo/Desktop/찐찐행정동420.csv',index=False)
```
