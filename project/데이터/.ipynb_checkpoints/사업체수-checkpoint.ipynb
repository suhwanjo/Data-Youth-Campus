# 8차 대분류

```python
# 칼럼명 지정
data_list = []
column = ['기준년도','집계구코드','통계항목','통계값']

dir_sector_8 = "C:/Users/ASUS/Desktop/기타반출자료/B022. 서울시 00~14년 집계구별 사업체 통계/2. 파일데이터/2000~2014(2014집계구단위기준)/04사업체/사업체수/8차 대분류"

for file in os.listdir(dir_sector_8):
    file_path = os.path.join(dir_sector_8, file)
    df = pd.read_csv(file_path, encoding='utf-8-sig',sep='^',names=column)
    data_list.append(df)
company = pd.concat(data_list, ignore_index=True)
```

```python
company
```

```python
# 지역 코드
gucode = pd.read_csv("C:/Users/ASUS/Desktop/데이터/기타반출자료/B022. 서울시 00~14년 집계구별 사업체 통계/2. 파일데이터/0. 지역코드_매칭표/집계구(서울)_지역코드_매칭표.txt")
gucode
```

```python
# Step 3: Drop duplicate codes
df_unique_codes = gucode.drop_duplicates(subset='ADSTRD_CD')

# Step 4: Group by code and check if values are the same
df_unique_codes['consistent_value'] = gucode.groupby('ADSTRD_CD')['ADSTRD_NM'].transform('nunique') == 1

# Step 5: Select rows with consistent values
unique_pairs = df_unique_codes[df_unique_codes['consistent_value']][['ADSTRD_CD', 'ADSTRD_NM']].reset_index(drop=True)

# Print the unique code-value pairs
print(unique_pairs)

unique_pairs.to_csv("행정동코드7자리_422.csv",index=False)
```

```python
dong_dic = {
    '금호2.3가동':'금호2·3가동',
    '면목3.8동':'면목3·8동',
    '상계3.4동':'상계3·4동',
    '상계6.7동':'상계6·7동',
    '종로1.2.3.4가동':'종로1·2·3·4가동',
    '종로5.6가동':'종로5·6가동',
    '중계2.3동':'중계2·3동'
}
# Function to apply the mapping
def apply_mapping(value):
    if value in dong_dic:
        return dong_dic[value]
    return value
```

```python
dong['ADSTRD_NM'] = dong['ADSTRD_NM'].apply(apply_mapping)
```

```python
total_dong = pd.merge(unique_pairs,dong,how='inner',on='ADSTRD_NM')
```
