```python
import pandas as pd
import numpy as np
import os
```

# 지역코드 매칭표

```python
# 지역 코드
gucode = pd.read_csv("데이터/기타반출자료/B022. 서울시 00~14년 집계구별 사업체 통계/2. 파일데이터/0. 지역코드_매칭표/집계구(서울)_지역코드_매칭표.txt")
gucode
```

```python
len(gucode['ADSTRD_NM'].unique()) # 행정동 수 422개
```

# 사용할 칼럼 지정 + 지역코드 매칭표와 결합

```python
data_list2 = []
# 사용할 칼럼 지정
column = ['기준년도','집계구코드','통계항목','통계값']

dir_sector_9 = "데이터/기타반출자료/B022. 서울시 00~14년 집계구별 사업체 통계/2. 파일데이터/2000~2014(2014집계구단위기준)/04사업체/종사자수/9차 대분류"
# 전체 년도를 하나의 데이터프레임에 저장
for file in os.listdir(dir_sector_9):
    file_path = os.path.join(dir_sector_9, file)
    df2 = pd.read_csv(file_path, encoding='utf-8-sig',sep='^',names=column)
    data_list2.append(df2)
company2 = pd.concat(data_list2, ignore_index=True)
company2
```

```python
company_gu2 = pd.merge(company2, gucode, how='inner', left_on='집계구코드', right_on='TOT_OA_CD')
company_gu2 = company_gu2[['기준년도','통계항목','통계값','SIGNGU_NM','ADSTRD_NM','ADSTRD_CD']]
company_gu2
```

# 결측치 제거

```python
# 결측치 제거
company_gu2.dropna(inplace=True)
company_gu2
```

```python
company_gu2['통계항목'].unique()
```

# 통계항목 매칭표와 결합

```python
code_sector_mapping_9 = {
    "cp_bem_001": "농업, 임업 및 어업",
    "cp_bem_002": "광업",
    "cp_bem_003": "제조업",
    "cp_bem_004": "전기, 가스, 증기 및 수도사업",
    "cp_bem_005": "하수 처리, 폐기물 재활용 및 환경 복원",
    "cp_bem_006": "건설업",
    "cp_bem_007": "도매 및 소매",
    "cp_bem_008": "운수업",
    "cp_bem_009": "숙박 및 음식점업",
    "cp_bem_010": "방송통신 및 정보서비스업",
    "cp_bem_011": "금융 및 보험업",
    "cp_bem_012": "부동산 및 임대업",
    "cp_bem_013": "전문, 과학 및 기술 서비스업",
    "cp_bem_014": "사업시설 관리 및 사업 지원 서비스업",
    "cp_bem_015": "공공 행정, 국방 및 사회보장 행정",
    "cp_bem_016": "교육 서비스업",
    "cp_bem_017": "보건 및 사회복지 서비스업",
    "cp_bem_018": "예술, 스포츠 및 여가관련 서비스업",
    "cp_bem_019": "협회 및 단체, 수리 및 기타 개인 서비스업",
    "cp_bem_999": "데이터 집계 없음",
}
```

```python
# 통계항목 매칭표 적용
company_gu2['통계항목'] = company_gu2['통계항목'].map(code_sector_mapping_9)
company_gu2
```

```python
# 데이터 집계가 없는 데이터는 제거
company_gu2 = company_gu2[company_gu2['통계항목']!='데이터 집계 없음']
company_gu2
```

```python
company_gu2['기준년도'].unique()
```

```python
# 2014년도 데이터만 사용
company_gu2 = company_gu2[company_gu2['기준년도']==2014]
company_gu2
```

```python
len(company_gu2['ADSTRD_NM'].unique()) # 상일동, 일원2동
```

# 행정동 맞추기

```python
# 상일동, 일원2동 제거
company_gu3 = company_gu2[~((company_gu2['ADSTRD_NM']=='상일동')|(company_gu2['ADSTRD_NM']=='일원2동'))]
company_gu3
```

```python
len(company_gu3['ADSTRD_NM'].unique()) # 신사동 때문에 이름은 420개
```

```python
len(company_gu3['ADSTRD_CD'].unique()) 
```
