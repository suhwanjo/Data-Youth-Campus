# 사업체 조사 결과 집계

```python
df = pd.read_csv("데이터/2015_서울시_사업체.csv.")
df
# 칼럼 설명
# ESTM_RPRS_SD_CD : 사업체대표자성별코드 ( 1:남자,2:여자 )
# ESTM_RPRS_APS : 사업체대표자연령 ( 1:20대이하,2:30대,3:40대,4:50대,5:60대,6:70대이상 )
# FND_YEAR : 창설년도, 월
# ORGFM_DIV_CD : 조직형태구분코드 ( 1:개인사업체,2:회사법인,3:회사이외법인,4:비법인단체,5:국가지자체)
# ORGFM_ESTM_DIV_CD : 조직형태사업체구분코드
# MBZ_INDST_CLS_CD : 주사업산업분류코드 -> 사업체,종사자수 데이터에 있는 9차 대분류임
```

```python
df.isnull().sum()
```

```python
df
```

```python
len(df['AD_CD'].unique())
```

```python
# 행정동별 사업체대표자성별을 집계
pivot_df = df.groupby(['AD_CD', 'ESTM_RPRS_SD_CD']).size().unstack(fill_value=0)
pivot_df.columns = [f'사업체대표자성별_{col}' for col in pivot_df.columns]
pivot_df = pivot_df.reset_index()

# 사업체 총 합을 계산해서 새로운 칼럼 생성
pivot_df['사업체_총합'] = pivot_df.iloc[:, 1:].sum(axis=1)
pivot_df
```

```python
# 칼럼 이름 변경
pivot_df = pivot_df.rename(columns={'사업체대표자성별_1':'대표자_남자','사업체대표자성별_2':'대표자_여자'})
```

```python
# 행정동별 사업체대표자연령대 수를 집계
pivot_df2 = df.groupby(['AD_CD', 'ESTM_RPRS_APS']).size().unstack(fill_value=0)
pivot_df2.columns = [f'사업체대표자연령대_{col}' for col in pivot_df2.columns]

pivot_df2 = pivot_df2.reset_index()
pivot_df2
```

```python
pivot_df2 = pivot_df2.rename(columns={'사업체대표자연령대_1':'대표자_20대이하','사업체대표자연령대_2':'대표자_30대','사업체대표자연령대_3':'대표자_40대','사업체대표자연령대_4':'대표자_50대','사업체대표자연령대_5':'대표자_60대','사업체대표자연령대_6':'대표자_70대이상'})
```

```python
# 행정동별 조직형태(개인사업체,회사법인 등) 수를 집계
pivot_df3 = df.groupby(['AD_CD', 'ORGFM_DIV_CD']).size().unstack(fill_value=0)

pivot_df3.columns = [f'조직형태구분_{col}' for col in pivot_df3.columns]

pivot_df3 = pivot_df3.reset_index()
pivot_df3
```

```python
# 칼럼 이름 변경
pivot_df3 = pivot_df3.rename(columns={'조직형태구분_1':'조직형태_개인사업체','조직형태구분_2':'조직형태_회사법인','조직형태구분_3':'조직형태_회사이외법인','조직형태구분_4':'조직형태_비법인단체','조직형태구분_5':'조직형태_국가지자체'})
```

```python
# 모든 집계 병합
df = pd.merge(pivot_df,pivot_df2,how='inner',on='AD_CD')
df = pd.merge(df,pivot_df3,how='inner',on='AD_CD')
df
```

```python
df.to_csv('데이터/사업체조사결과.csv',index=False)
```

```python
# 사용하지 않는 동은 제거
no_dong = [
 '개봉본동',
 '공릉1.3동',
 '구로본동',
 '길1동',
 '남대문로5가동',
 '노고산동',
 '논현동',
 '도곡동',
 '도화1동',
 '도화2동',
 '독산1동분소',
 '명륜3가동',
 '상일동',
 '상일제1동',
 '상일제2동',
 '신당1동',
 '신당2동',
 '신당3동',
 '신당4동',
 '신당6동',
 '신사동',
 '영등포1동',
 '월계4동',
 '위례동',
 '을지로3.4.5가동',
 '일원2동',
 '태평로1가동',
 '하일동',
 '항동']

df = df[~df["행정동-연료별 분류"].isin(no_dong)]
df
```

```python
dong_dic = {
    '금호2.3가동':'금호2·3가동',
    '면목3.8동':'면목3·8동',
    '상계3.4동':'상계3·4동',
    '상계6.7동':'상계6·7동',
    '종로1.2.3.4가동':'종로1·2·3·4가동',
    '종로5.6가동':'종로5·6가동',
    '중계2.3동':'중계2·3동'
}
# .을 ·으로 변경
def apply_mapping(value):
    if value in dong_dic:
        return dong_dic[value]
    return value
df['행정동-연료별 분류'] = df['행정동-연료별 분류'].apply(apply_mapping)
```
